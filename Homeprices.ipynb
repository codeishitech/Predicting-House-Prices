```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("homeprices.csv")
df.head()
```

```python
df.bedrooms.median()
from math import floor
df.bedrooms=df.bedrooms.fillna(floor(df.bedrooms.median()))

df.head()
```

```python
df.columns=df.columns.str.strip()
```

```python
reg=LinearRegression()
x=df[["area","bedrooms","age"]]   #x is assigned/contains the 2d array or dataframe containing the features, this is 2d because it has multiple features
y=df["price"]                # y contains the series price that is to be predicted(since it's a single value)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
reg.fit(x_train,y_train)
y_predict=reg.predict(x_test)
```

```python
print("m=",reg.coef_) #m
print("b=",reg.intercept_)
```
